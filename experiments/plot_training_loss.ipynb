```python
%matplotlib inline
import matplotlib.pyplot as plt

import pickle 
import numpy as np
import pandas as pd
import os
```

```python
root = 'results/mnist_5/'
filename = root + 'experiment_results.pkl'
with open(filename, 'rb') as f:
    tab_results = pickle.load(f)
```

```python
def unpack_training_results(tab_results):
    checkpoints = [res[0] for res in tab_results]
    loss = [res[1] for res in tab_results]
    return checkpoints, loss
```

```python
checkpoints, loss = unpack_training_results(tab_results)
```

```python
root = 'results/mnist_6/'
filename = root + 'experiment_results.pkl'
with open(filename, 'rb') as f:
    tab_results_2 = pickle.load(f)
```

```python
checkpoints_2, loss_2 = unpack_training_results(tab_results_2)
```

```python
root = 'results/mnist_7/'
filename = root + 'experiment_results.pkl'
with open(filename, 'rb') as f:
    tab_results_3 = pickle.load(f)
```

```python
checkpoints_3, loss_3 = unpack_training_results(tab_results_3)
```

```python
f, axarr = plt.subplots(1, 1, sharex='col', sharey='row', figsize=(7,5), dpi=150)
plt.plot(checkpoints, loss, label='Simple net')
plt.plot(checkpoints_2, loss_2, label='Graph net')
plt.plot(checkpoints_3, loss_3, label='Graph net (old)')
plt.grid()
plt.xlabel('Iteration')
plt.ylabel('Training loss')
plt.legend()
```
